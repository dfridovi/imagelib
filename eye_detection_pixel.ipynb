# SVM-based Eye Detection
## Use an SVM to detect eyes, given a training image and mask.

```python
import BasicFunctions as bf
import numpy as np

%matplotlib inline
```

```python
img = bf.imread("obama.jpg")
bf.imshow(img)
```

```python
train = img[200:225, 480:525]
bf.imshow(train)
```

```python
mask_pos = np.zeros(train.shape, dtype=np.bool)
mask_pos[15:20, 15:35] = True
mask_pos[5:15, 30:40] = True
num_pos = 5 * 20 + 10 * 10

mask_neg = np.zeros(img.shape, dtype=np.bool)
mask_neg[500:, :] = True
num_neg = (mask_neg.shape[0] - 500) * mask_neg.shape[1]

print (num_pos, num_neg)
```

```python
from sklearn import svm
from sklearn import preprocessing
```

```python
training_pos = train[mask_pos].reshape((num_pos, 3))
training_neg = img[mask_neg].reshape((num_neg, 3))
training_set = np.vstack((training_pos, training_neg))

training_labels = np.ones(num_pos + num_neg)
training_labels[num_pos:] = 0

scaler = preprocessing.StandardScaler().fit(training_set)
training_set = scaler.transform(training_set)
```

```python
weights = {0 : 1.0, 1 : 5.0}
clf = svm.SVC(C=10.0, gamma=0.1, kernel="rbf", class_weight=weights)
clf.fit(training_set, training_labels)
```

```python
output_image = np.copy(img).reshape((img.shape[0]*img.shape[1], 3))

testing_set = img.reshape((img.shape[0]*img.shape[1], 3))
testing_set = scaler.transform(testing_set)

output_labels = clf.predict(testing_set)
output_image[output_labels == 0] = 0.5
output_image = output_image.reshape(img.shape)

bf.imshow(output_image)
```
